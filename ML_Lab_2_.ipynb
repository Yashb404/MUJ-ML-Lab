```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

df.loc[0, 'math score'] = np.nan
df.loc[1, 'gender'] = np.nan
df.loc[2, 'math score'] = 999

print(" Experiment 1 ")
print(df.isnull().head())
print(df.notnull().head())
print(df.isnull().sum())

print("\n Experiment 2 ")
print(df.shape)
df_clean = df.dropna()
print(df_clean.shape)

print("\n Experiment 3 ")
df_filled = df.copy()
df_filled['math score'] = df_filled['math score'].fillna(df_filled['math score'].median())
df_filled['gender'] = df_filled['gender'].fillna('Unknown')
print(df_filled.isnull().sum())

print("\n Experiment 4 ")
print(df_filled.dtypes)
df_filled['math score'] = df_filled['math score'].astype('float64')
print(df_filled.dtypes)

print("\n Experiment 5 ")
df_renamed = df_filled.rename(columns={'math score': 'Math_Score', 'reading score': 'Reading_Score'})
print(df_renamed.head())

print("\n Experiment 6 ")
df_corrected = df_renamed.replace(999, df_renamed['Math_Score'].median())
print(df_corrected.head())

print("\n Experiment 7 ")
plt.plot(df_corrected.index[:20], df_corrected['Math_Score'][:20])
plt.show()
category_counts = df_corrected['gender'].value_counts()
plt.bar(category_counts.index, category_counts.values)
plt.show()

print("\n Experiment 8 ")
plt.scatter(df_corrected['Math_Score'], df_corrected['Reading_Score'])
plt.show()

print("\n Experiment 9 ")
plt.hist(df_corrected['Math_Score'], bins=15)
plt.show()

print("\n Experiment 10 ")
plt.boxplot(df_corrected['Math_Score'])
plt.show()
```

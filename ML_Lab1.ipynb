```python
import numpy as np
import pandas as pd

url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"

print(" Experiment 1 ")
list_data = [10, 20, 30, 40, 50]
arr1 = np.array(list_data)
print(arr1)
print(arr1.shape)
print(arr1.dtype)

print("\n Experiment 2 ")
zeros_mat = np.zeros((3, 4))
ones_mat = np.ones((2, 5))
print(zeros_mat)
print(zeros_mat.shape)
print(ones_mat)
print(ones_mat.shape)

print("\n Experiment 3 ")
arr_1_to_12 = np.arange(1, 13)
reshaped_arr = arr_1_to_12.reshape(3, 4)
print(arr_1_to_12)
print(arr_1_to_12.shape)
print(reshaped_arr)
print(reshaped_arr.shape)

print("\n Experiment 4 ")
rand_arr = np.random.randint(1, 100, 10)
print(rand_arr)
print(rand_arr.min())
print(rand_arr.max())
print(rand_arr.mean())
print(rand_arr.sum())

print("\n Experiment 5 ")
data_dict = {'Name': ['Alice', 'Bob', 'Charlie'], 'Marks': [85, 90, 78]}
df_students = pd.DataFrame(data_dict)
print(df_students)
print(df_students.dtypes)

print("\n Experiment 6 ")
df_csv = pd.read_csv(url)
print(df_csv.head())
print(df_csv.tail())

print("\n Experiment 7 ")
df_csv.info()

print("\n Experiment 8 ")
summary = df_csv.describe()
print(summary)
print("\nMean:\n", summary.loc['mean'])
print("\nStandard Deviation:\n", summary.loc['std'])
print("\nMinimum:\n", summary.loc['min'])
print("\nMaximum:\n", summary.loc['max'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

print(" Experiment 1 ")
print(df.head())

num_cols = df.select_dtypes(include=np.number)
cat_cols = df.select_dtypes(exclude=np.number)

print("\nNumerical Columns:\n", num_cols.columns.tolist())
print("\nCategorical Columns:\n", cat_cols.columns.tolist())

print("\nMean:\n", num_cols.mean())
print("\nMedian:\n", num_cols.median())
print("\nMode:\n", num_cols.mode().iloc[0])

print("\nMinimum:\n", num_cols.min())
print("\nMaximum:\n", num_cols.max())
print("\nSum:\n", num_cols.sum())
print("\nVariance:\n", num_cols.var())
print("\nStandard Deviation:\n", num_cols.std())

print("\n Experiment 2 ")
print("\nQ1 (25th Percentile):\n", num_cols.quantile(0.25))
print("\nQ2 (Median / 50th Percentile):\n", num_cols.quantile(0.50))
print("\nQ3 (75th Percentile):\n", num_cols.quantile(0.75))

print("\n Experiment 3 ")
print("\nCorrelation Coefficients:\n", num_cols.corr())
print("\nCovariance:\n", num_cols.cov())

print("\n Experiment 4 ")
num_cols.hist(bins=15, figsize=(10, 5), edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()

plt.figure(figsize=(8, 5))
num_cols.boxplot()
plt.title('Boxplots of Numerical Features')
plt.show()
```
